# Тепер рисуем остальную сову

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
%matplotlib inline # магия для встроенного рисования в тетрадке
````

Сначала запилим свой датасет на котором это всё крутить и заодно порисуем в matplotlib

````python
np.random.seed(1)
x1 = np.random.normal(loc = 1.0, scale = 0.5, size = 100) # loc это медиана, scale - сигма
x2 = np.random.normal(loc = -1.0, scale = 1.5, size = 100)
y1 = np.random.normal(loc = 2.0, scale = 1.5, size = 100)
y2 = np.random.normal(loc = 3.0, scale = 0.5, size = 100)
````

````python
plt.figure(figsize = (8,8)) # размер графика
plt.xlim(-5, 4) # границы графика
plt.ylim(-2, 6)
plt.plot(x1, y1, 'r.', label = 'class 0')
plt.plot(x2, y2, 'b.', label = 'class 1')
plt.legend(loc = 0)
````

Надо теперь сделать датасет, который можно будет кормить нашим классификаторам в понятном им формате

````python
X = np.vstack([np.array([x1,y1]).T, np.array([x2,y2]).T]) # склеиваем в двухмерный массив с двумя столбцами
````

````python
y = np.hstack([np.zeros(100), np.ones(100)])
````

Здесь мы создаем нужный классификатор с нужными параметрами, а потом вызываем у него метод .fit, который тренирует классификатор на переданном сете с заданными ответами

````python
dt1 = DecisionTreeClassifier(max_depth = 1, random_state = 1) # дерево глубины 1
dt1.fit(X, y)
dt2 = DecisionTreeClassifier(max_depth = 2, random_state = 1) # дерево глубины 2
dt2.fit(X, y)
dt3 = DecisionTreeClassifier(max_depth = 3, random_state = 1) # дерево глубины 3
dt3.fit(X, y)
dtn = DecisionTreeClassifier(random_state = 1) # дерево, строящееся до полного разделения датасета
dtn.fit(X, y)
````

Ничего не понятно, надо что-нибудь нарисовать красивое, тем более у нас размерность данных - 2, можно прям на плоскости рисовать и смотреть границы решений деревьев

````python
def plot_decision(X, y, model):
    plt.figure(figsize = (8, 6))
    
    plot_step = 0.005  # fine step width for decision surface contours
    plot_step_coarser = 0.5
    plot_colors = "rb"
    cmap = plt.cm.RdYlBu

    # Now plot the decision boundary using a fine mesh as input to a
    # filled contour plot
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                         np.arange(y_min, y_max, plot_step))

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    cs = plt.contourf(xx, yy, Z, cmap=cmap)

    # Build a coarser grid to plot a set of ensemble classifications
    # to show how these are different to what we see in the decision
    # surfaces. These points are regularly space and do not have a black outline
    #xx_coarser, yy_coarser = np.meshgrid(np.arange(x_min, x_max, plot_step_coarser),
    #                                     np.arange(y_min, y_max, plot_step_coarser))
    #Z_points_coarser = model.predict(np.c_[xx_coarser.ravel(), yy_coarser.ravel()]).reshape(xx_coarser.shape)
    #cs_points = plt.scatter(xx_coarser, yy_coarser, s=15, c=Z_points_coarser, cmap=cmap, edgecolors="none")

    # Plot the training points, these are clustered together and have a
    # black outline
    for i, c in zip(xrange(2), plot_colors):
        idx = np.where(y == i)
        plt.scatter(X[idx, 0], X[idx, 1], c=c,
                    cmap=cmap)
````

````python
plot_decision(X, y, dt1) # рисуем решение деревьев
````

````python
plot_decision(X, y, dt2)
````

````python
plot_decision(X, y, dt3)
````

````python
plot_decision(X, y, dtn)
````

Казалось бы - чем больше глубина дерева, тем лучше мы приближаем данные и получается счастье, радость, жвачка, но нет.

Фактически мы получаем оверфит и можно посмотреть как всё плохо, если нагенерить тестовых данных из тех же распределений и посчитать на них ROC AUC (крутая метрика, позволяющая оценивать качество классификации)

### Лирическое отступление про ROC AUC
ROC AUC - receiver operator characteristic area under the curve

Эта штука - площадь под графиком кривой, которая оценивает отношение true positive rate к false positive rate. Ей очень удобно мерить качество бинарной классификации для разных моделей и сравнивать их точность.

Фактически ответ классификатора это вероятность принадлежности к нужному классу. Для итоговой бинарной классификации нам нужно поставить на интервале [0, 1] порог, такой что всё что больше него будет классифицировано как 1, а ниже - 0. Но поскольку от выбора этого порога очень сильно зависит количество ошибок первого и второго рода и всё счастье, которым меряется бинарная классификация, а у нас тут вероятности и вообще, нам хотелось бы сделать метрику, которая не зависит от этого порога.

ROC AUC как раз такой. Это отношение tpr к fpr, когда мы двигаем этот порог от 0 до 1.

````python
# нагенерим тестовый датасет с теми же параметрами
np.random.seed(2)
x1 = np.random.normal(loc = 1.0, scale = 0.5, size = 50)
x2 = np.random.normal(loc = -1.0, scale = 1.5, size = 50)
y1 = np.random.normal(loc = 2.0, scale = 1.5, size = 50)
y2 = np.random.normal(loc = 3.0, scale = 0.5, size = 50)
````

````python
X_test = np.vstack([np.array([x1,y1]).T, np.array([x2,y2]).T])
````

````python
y_test = np.hstack([np.zeros(50), np.ones(50)])
````

````python
from sklearn.metrics import roc_auc_score
````

````python
print 'tree1 roc auc =', roc_auc_score(y_test, dt1.predict_proba(X_test)[:,1])
print 'tree2 roc auc =', roc_auc_score(y_test, dt2.predict_proba(X_test)[:,1])
print 'tree3 roc auc =', roc_auc_score(y_test, dt3.predict_proba(X_test)[:,1])
print 'treen roc auc =', roc_auc_score(y_test, dtn.predict_proba(X_test)[:,1])
````

Пример самой ROC кривой

````python
from sklearn import metrics
from sklearn.metrics import roc_curve, auc
plt.figure(figsize = (6, 6))
trees = [dt1, dt2, dt3, dtn]
for i in xrange(len(trees)):
    pred = trees[i].predict_proba(X_test)[:,1]
    fpr, tpr, thresholds = roc_curve(y_test, pred)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, lw=1, label='ROC %s (area = %0.3f)' % (i+1, roc_auc))
plt.plot([0, 1], [0, 1], '--', color=(0.6, 0.6, 0.6), label='Luck')
plt.xlim([-0.05, 1.05])
plt.ylim([-0.05, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic example')
plt.legend(loc="lower right")
````

## А теперь попробуем запилить DummyForest
Это будет что-то в духе RandomForest - модель ансамбля, объединяющая несколько деревьев решений, чтобы сделать модель более робастной и не зависящей от флуктуаций данных.

Для этого нужно тренировать каждое из деревьев леса не на всем датасете, а на случайной выборке из него.

Нам нужно, чтобы у него были методы ```__init__()``` (куда же без него), ```fit()``` (тренировать деревья), ```predict_proba()``` (для предсказания вероятностей) и ```predict()``` (для предсказания бинарных ответов 0 или 1). И параметры ```n_estimators``` - для числа деревьев и ```estimators_``` - сам список деревьев, чтобы их рисовать.

Методы ансамблирования можно использовать какие хочется, но самый простой - усреднение решений составляющих деревьев.

````python
# для рисования области решений леса
def plot_decision_forest(X, y, model):
    plt.figure(figsize = (8, 6))
    
    plot_step = 0.005  # fine step width for decision surface contours
    plot_step_coarser = 0.5
    plot_colors = "rb"
    cmap = plt.cm.RdYlBu

    # Now plot the decision boundary using a fine mesh as input to a
    # filled contour plot
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                         np.arange(y_min, y_max, plot_step))

    # Choose alpha blend level with respect to the number of estimators
    # that are in use (noting that AdaBoost can use fewer estimators
    # than its maximum if it achieves a good enough fit early on)
    estimator_alpha = 1.0 / len(model.estimators_)
    for tree in model.estimators_:
        Z = tree.predict(np.c_[xx.ravel(), yy.ravel()])
        Z = Z.reshape(xx.shape)
        cs = plt.contourf(xx, yy, Z, alpha=estimator_alpha, cmap=cmap)

    # Build a coarser grid to plot a set of ensemble classifications
    # to show how these are different to what we see in the decision
    # surfaces. These points are regularly space and do not have a black outline
    xx_coarser, yy_coarser = np.meshgrid(np.arange(x_min, x_max, plot_step_coarser),
                                         np.arange(y_min, y_max, plot_step_coarser))
    Z_points_coarser = model.predict(np.c_[xx_coarser.ravel(), yy_coarser.ravel()]).reshape(xx_coarser.shape)
    cs_points = plt.scatter(xx_coarser, yy_coarser, s=15, c=Z_points_coarser, cmap=cmap, edgecolors="none")

    # Plot the training points, these are clustered together and have a
    # black outline
    for i, c in zip(xrange(2), plot_colors):
        idx = np.where(y == i)
        plt.scatter(X[idx, 0], X[idx, 1], c=c,
                    cmap=cmap)
````

````python
clf = DummyForest(n_estimators = 5)
clf.fit(X, y)
predictions = clf.predict_proba(X_test)[:,1]
print 'ROC AUC =', roc_auc_score(y_test, predictions)
plot_decision_forest(X, y, clf)
````
